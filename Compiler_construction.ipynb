**LEX TOOL**

Lex receives a lexical syntax description from file.l in the form of regular expressions and actions. Lex reads the description from standard input if file.l is not provided or if the file is called -. (stdin). In order to scan such expressions, it generates a set of tables that, along with other prototype code from /etc/yylex.c, form a lexical analyzer. The recognized produced can be used with yacc.
